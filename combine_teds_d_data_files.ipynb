```python
import pandas as pd
import numpy as np
import os
```

```python
directory = "C:\\Users\\ericm\\OneDrive\\Documents\\MADS\\Capstone\\data\\"

dataframes = []

all_columns = set()

# iterate over the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'): 
        # combine the directory path and filename to get the full path to the CSV file
        file_path = os.path.join(directory, filename)

        # read the CSV file into a pandas DataFrame
        df = pd.read_csv(file_path)

        # print the shape of the DataFrame
        print(f'Shape of DataFrame from {filename}: {df.shape}')

        # append the DataFrame to the list
        dataframes.append(df)

        # add column names to the set of all column names
        all_columns.update(df.columns)

    # elif filename.endswith('.tsv'):
    #     # combine the directory path and filename to get the full path to the TSV file
    #     file_path = os.path.join(directory, filename)

    #     # read the TSV file into a pandas DataFrame
    #     df = pd.read_csv(file_path, sep='\t')

    #     # print the shape of the DataFrame
    #     print(f'Shape of DataFrame from {filename}: {df.shape}')

    #     # append the DataFrame to the list
    #     dataframes.append(df)

    #     # add column names to the set of all column names
    #     all_columns.update(df.columns)
        
all_missing_columns = set()

# find columns that are not shared by all dataframes
for i, df in enumerate(dataframes):
    missing_columns = all_columns.difference(df.columns)
    if missing_columns:
        print(f'DataFrame {i} is missing the following columns: {missing_columns}')
        all_missing_columns.update(missing_columns)
```

```python
# iterate over dataframes and delete columns that are not shared by all dataframes
for i, df in enumerate(dataframes):
    for col in all_missing_columns:
        if col in df.columns:
            df.drop(columns=col, inplace=True)
```

```python
for df in dataframes:
    print(df.shape)
```

```python
# concatenate all the dataframes
combined_df = pd.concat(dataframes, ignore_index=True)

print(combined_df.shape)
combined_df.head()
```

```python
# keep only features of interest

feature_to_keep = [
    'DISYR',
    'AGE',
    'GENDER',
    'RACE',
    'MARSTAT',
    'EDUC',
    'EMPLOY',
    'EMPLOY_D',
    'LIVARAG',
    'LIVARAG_D',
    'ARRESTS',
    'ARRESTS_D',
    'SERVICES',
    'SERVICES_D',
    'REASON',
    'LOS',
    'PSOURCE',
    'NOPRIOR',
    'SUB1',
    'FREQ1',
    'FRSTUSE1',
    'ALCDRUG',
    'DSMCRIT',
    'PSYPROB',
    'HLTHINS',
    'PRIMPAY',
    'FREQ_ATND_SELF_HELP'
]

combined_df = combined_df[feature_to_keep]

print(combined_df.shape)
combined_df.head()
```

```python
# write the combined DataFrame to a CSV file
combined_df.to_csv('C:\\Users\\ericm\\OneDrive\\Documents\\MADS\\Capstone\\data\\combined_teds_d_data.csv', index=False)
```
