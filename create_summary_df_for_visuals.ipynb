```python
import pandas as pd
```

```python
# Create an empty dataframe to hold the results
summary_df = pd.DataFrame()

df = pd.read_csv('cleaned_data.csv')

# Iterate over each column in the original dataframe
for column in df.columns:
    # Count the number of occurrences of each unique value in the column
    value_counts = df[column].value_counts()
    # Convert the Series to a DataFrame and transpose it so that the unique
    # values become the columns
    value_counts_df = value_counts.to_frame().T
    # Append the result to the summary dataframe
    summary_df = summary_df.append(value_counts_df)

# Display the first few rows of the summary dataframe
summary_df.head()

summary_df.to_csv('cleaned_data_counts.csv')
```
